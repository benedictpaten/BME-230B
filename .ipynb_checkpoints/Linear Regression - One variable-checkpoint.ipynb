```python
# LINEAR REGRESSION - PART 1 (ONE VARIABLE)
# DATASET http://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/
import numpy as np
import pandas as pd
```

```python
# Read input
inputdata = pd.read_csv("./data/breast-cancer-wisconsin.data.csv")
inputdata.head(3)
```

```python
#for logisitic regression we need numbers instead of letters
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()

for col in inputdata.columns:
    inputdata[col] = encoder.fit_transform(inputdata[col])
    
inputdata.head(10)
```

```python
def predict(x, w, b):
    return x*w + b
```

```python
def cost_function(x, y, w, b):
    features = len(x)
    total_error = 0.0
    for i in range(features):
        total_error += (y[i] - (w*x[i] + b))**2
    return total_error / features
```

```python
def update_weights(x, y, w, b, learning_rate):
    weight_deriv = 0
    bias_deriv = 0
    features = len(x)

    for i in range(features):
        # Calculate partial derivatives
        # -2x(y - (mx + b))
        weight_deriv += -2*x[i] * (y[i] - (w*x[i] + b))

        # -2(y - (mx + b))
        bias_deriv += -2*(y[i] - (w*x[i] + b))

    # We subtract because the derivatives point in direction of steepest ascent
    w -= (weight_deriv / features) * learning_rate
    b -= (bias_deriv / features) * learning_rate

    return w, b
```

```python
def train_model(x, y, w, b, learning_rate, iters):
    cost_history = []

    for i in range(iters):
        w,b = update_weights(x, y, w, b, learning_rate)

        #Calculate cost for auditing purposes
        cost = cost_function(x, y, w, b)
        cost_history.append(cost)

        # Log Progress
        if i % 10 == 0:
            print("iter: ", str(i), " cost: ",  str(cost))

    return w, b, cost_history
```

```python
def sigmoid(z):
  return 1.0 / (1 + np.exp(-z))
```

```python
x = inputdata['mitoses']
y = inputdata['class']
w, b, cost_history = train_model(x, y, 0.01, 0.01, 0.02, 500)
print(w, b)
### SCORE ###
logits_0 = []
logits_1 = []
for i in range(len(x)):
    p = predict(x[i], w, b)
    if y[i] == 0:
        logits_0.append(p)
    else:
        logits_1.append(p)
# print(logits_0)
print(sum(logits_1)/len(logits_1) - sum(logits_0)/len(logits_0))
import matplotlib.pyplot as plt
n, bins, patches = plt.hist(logits_0, normed=True, cumulative=True, bins=500, facecolor='green', alpha=0.75)
plt.axis([0,1,0,1])
plt.show()
n, bins, patches = plt.hist(logits_1, normed=True, cumulative=True, bins=500, facecolor='red', alpha=0.75)
plt.axis([0,1,0,1])
plt.show()
### SCORE END ###
```

```python
x = [[x] for x in inputdata['cap-shape'].as_matrix()]
y = [[y] for y in inputdata['poisonous'].as_matrix()]
from sklearn import datasets, linear_model
lm_model = linear_model.LinearRegression(fit_intercept=True, normalize=True, copy_X=True, n_jobs=1)
lm_model.fit(x, y)
print("Co-efficent: ", lm_model.coef_)
print("Intercept: ", lm_model.intercept_)
print(lm_model.score(x,y))

# So, we have learned the same co-efficents (or, almost the same) through a library.
```

```python
features = inputdata.keys()
feature_scores = []

for feature in features:
    x = [[x] for x in inputdata[feature].as_matrix()]
    y = [[y] for y in inputdata['poisonous'].as_matrix()]
    lm_model = linear_model.LinearRegression(fit_intercept=True, normalize=True, copy_X=True, n_jobs=1)
    lm_model.fit(x, y)
    feature_scores.append([feature, lm_model.score(x,y)])
feature_scores.sort(key=lambda x: x[1], reverse=True)
for feature_score in feature_scores:
    print(feature_score)
```
